```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
# prompt: import simple imputer

import numpy as np
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

# impute missing values in df['age']
df['age'] = imputer.fit_transform(df['age'].to_numpy().reshape(-1,1))
```

```python
sns.get_dataset_names()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
x = df[['age','fare']]
y = df['survived']
```

```python
# prompt: train test split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print('Train data shape:', x_train.shape)
print('Test data shape:', x_test.shape)
```

```python
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
# prompt: import classification report,accuracy

from sklearn import metrics
from sklearn.metrics import accuracy_score, classification_report

print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
import numpy as np

# Original 1D array
data = np.array([80, 50])

# Reshaping the data to 2D array
X = data.reshape(1, -1)

# Making the prediction
prediction = model.predict(X)

print(prediction)
```

```python
# prompt: plot the model

import matplotlib.pyplot as plt

# Create a scatter plot of the data
plt.scatter(df['age'], df['survived'], color="green")

# Plot the regression line
plt.plot(x_test, model.predict(x_test), color='red')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Survived')
plt.title('Logistic Regression Model for Survival')

# Show the plot
plt.show()
```
